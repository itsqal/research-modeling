## Importing dataset

```python
df = pd.read_csv('../ews/data/dataset/draft_1_cp_2.csv', sep=',')
df_pattern = pd.read_csv('../ews/data/dataset/sequence_mining_output.csv', sep=',')

print("Datases loaded successfully.")
```

```python
print(df.info())
print('\n')
print(df_pattern.info())
```

```python
print(f"Shape of mai dataset: {df.shape}")
print(f"Shape of pattern dataset: {df_pattern.shape}")
```

```python
from ews.spm.pattern_filter import filter_patterns

filtered_patterns = filter_patterns(df_pattern, min_average_prob=80, prob_column='mean_prob')

print(f"{df_pattern.shape[0] - filtered_patterns.shape[0]} patterns filtered out. ")
```

## Combine pattern into data

Check for pattern in main data and match with filtered patterns

```python
from ews.spm.combine_pattern_to_data import combine_pattern_to_data

df_combined = combine_pattern_to_data(df, filtered_patterns, pattern_column='pattern', sequence_column='natural_signs_used')

print(f"Shape of combined dataset: {df_combined.shape}")
```

Save the combined dataframe

```python
df_combined.to_csv("../ews/data/dataset/draft_1_cp_3.csv")
```
